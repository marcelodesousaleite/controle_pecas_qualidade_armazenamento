```python
class Piece:
    def __init__(self, piece_id, weight, color, length):
        self.piece_id = piece_id
        self.weight = weight
        self.color = color
        self.length = length

    def __repr__(self):
        return f"Piece(ID: {self.piece_id}, Weight: {self.weight}g, Color: {self.color}, Length: {self.length}cm)"
```

A classe `Piece` representa uma peça fabricada com seus atributos: ID, peso, cor e comprimento.

```python
class QualityControlSystem:
    def __init__(self, box_capacity=10):
        self.box_capacity = box_capacity
        self.approved_pieces = []
        self.rejected_pieces = []
        self.current_box = []
        self.boxes_used = 0
        self.rejection_reasons = {}

    def _close_box(self):
        if self.current_box:
            # For simplicity, we just count boxes. In a real system, you might store box contents.
            self.boxes_used += 1
            print(f"Caixa {self.boxes_used} fechada com {len(self.current_box)} peças.")
            self.current_box = [] # Start a new empty box

    def process_piece(self, piece):
        approved = True
        rejection_reason = []

        # Weight check
        if not (95 <= piece.weight <= 105):
            approved = False
            rejection_reason.append(f"Peso ({piece.weight}g) fora da faixa (95g-105g).")

        # Color check
        if piece.color.lower() not in ['azul', 'verde']:
            approved = False
            rejection_reason.append(f"Cor ({piece.color}) não é azul ou verde.")

        # Length check
        if not (10 <= piece.length <= 20):
            approved = False
            rejection_reason.append(f"Comprimento ({piece.length}cm) fora da faixa (10cm-20cm).")

        if approved:
            self.approved_pieces.append(piece)
            self.current_box.append(piece)
            print(f"Peça {piece.piece_id} APROVADA. Adicionada à caixa.")
            if len(self.current_box) == self.box_capacity:
                self._close_box()
        else:
            self.rejected_pieces.append(piece)
            reason_str = "; ".join(rejection_reason)
            self.rejection_reasons[piece.piece_id] = reason_str
            print(f"Peça {piece.piece_id} REPROVADA. Motivo: {reason_str}")

    def generate_report(self):
        # Close any open box at the end of processing
        if self.current_box:
            self._close_box()

        print("\n--- Relatório de Produção e Qualidade ---")
        print(f"Total de peças processadas: {len(self.approved_pieces) + len(self.rejected_pieces)}")
        print(f"Total de peças APROVADAS: {len(self.approved_pieces)}")
        print(f"Total de peças REPROVADAS: {len(self.rejected_pieces)}")
        print(f"Quantidade de caixas utilizadas: {self.boxes_used}")

        if self.rejected_pieces:
            print("\nPeças Reprovadas e Motivos:")
            for piece in self.rejected_pieces:
                print(f"  - Peça ID: {piece.piece_id}, Motivo: {self.rejection_reasons.get(piece.piece_id, 'N/A')}")
        else:
            print("\nNenhuma peça foi reprovada.")

        print("------------------------------------------")
```

A classe `QualityControlSystem` gerencia todo o processo:

*   `__init__`: Inicializa o sistema com a capacidade da caixa e listas para peças aprovadas, reprovadas e a caixa atual.
*   `_close_box`: Método interno para fechar uma caixa quando ela atinge a capacidade.
*   `process_piece`: Avalia uma peça com base nos critérios de peso, cor e comprimento. Adiciona a peça à lista apropriada e gerencia o enchimento das caixas.
*   `generate_report`: Compila e exibe um relatório detalhado com totais de peças aprovadas/reprovadas, motivos de reprovação e o número de caixas utilizadas.

```python
# Exemplo de uso do sistema
qc_system = QualityControlSystem(box_capacity=3) # Capacidade de caixa reduzida para demonstração

# Gerando algumas peças de exemplo
pieces_to_process = [
    Piece("P001", 100, "azul", 15),   # Aprovada
    Piece("P002", 90, "azul", 15),    # Reprovada (peso)
    Piece("P003", 102, "verde", 12),  # Aprovada
    Piece("P004", 110, "azul", 18),   # Reprovada (peso)
    Piece("P005", 98, "vermelho", 16), # Reprovada (cor)
    Piece("P006", 97, "verde", 22),   # Reprovada (comprimento)
    Piece("P007", 99, "azul", 11),   # Aprovada
    Piece("P008", 101, "verde", 19),  # Aprovada
    Piece("P009", 103, "azul", 14),   # Aprovada
    Piece("P010", 100, "verde", 15),  # Aprovada
    Piece("P011", 96, "azul", 10),   # Aprovada
    Piece("P012", 104, "verde", 20),  # Aprovada
]

print("--- Iniciando Processamento das Peças ---")
for piece in pieces_to_process:
    qc_system.process_piece(piece)
print("--- Processamento Concluído ---")
```

Neste bloco, criamos uma instância do `QualityControlSystem` e processamos uma série de peças de exemplo, com diferentes características para testar os critérios de aprovação e reprovação. A capacidade da caixa foi definida como 3 para facilitar a visualização do fechamento das caixas.

```python
# Gerando o relatório final
qc_system.generate_report()
```

Por fim, este comando gera o relatório consolidado, mostrando os totais de peças aprovadas e reprovadas, os motivos de reprovação e o número total de caixas utilizadas.
